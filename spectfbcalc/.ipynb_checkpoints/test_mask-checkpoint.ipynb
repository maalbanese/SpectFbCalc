```python
import sys
import os
import glob

import numpy as np
import xarray as xr
import pickle

from climtools import climtools_lib as ctl
from matplotlib import pyplot as plt
import matplotlib.cbook as cbook
```

```python
import spectfbcalc_lib as sfc
```

```python
import cartopy.crs as ccrs
```

```python
path='/ec/res4/scratch/ccff/tunecs/abrupt_cmip6_r8/ta/*.nc'
ta=xr.open_mfdataset(path)
ta=sfc.check_vertical(ta)
ta
```

```python
p=ta.plev
p
```

```python
#costanti
g=9.81
cp=1005
R= 0.2870
k=R/cp
```

```python
n=ta.sizes['plev']
n
```

```python
result=[]
plevs=[]
for i in range (0, n-1):
    lev1_P=(p.isel(plev=i)).item()
    lev1_T=ta.sel(plev= lev1_P)
    lev2_P=(p.isel(plev=i+1)).item()
    lev2_T=ta.sel(plev= lev2_P)
    a=(lev2_T['ta']-lev1_T['ta'])
    b=((lev2_P*100)**k -(lev1_P*100)**k) #*100 bc datas are in hPa
    c=((lev1_P*100)**k+(lev2_P*100)**k)
    d=(lev1_T['ta']+lev2_T['ta'])
    lapse=(a/b)*(c/d)*(k*g/R)
    
    plevs.append(lev1_P)
    result.append(lapse)
lapse_da = xr.concat(result, dim="plev").assign_coords(plev=plevs)
cond = xr.where(lapse_da.plev < 100, lapse_da <= -2, True)

mask = cond.astype(int).cumprod("plev")

mask = mask.where(mask == 1)
```

```python
mask=sfc.mask_str(ta['ta'])
```

```python
mask
```

```python
ds_zonal = mask.sel(time=slice('1850-01-16T12:00:00', '1859-01-16T12:00:00')).mean(dim="lon")
ds_month = ds_zonal.groupby("time.month").mean("time")


fig, axes = plt.subplots(3, 4, figsize=(16,10), sharex=True, sharey=True)

for i, ax in enumerate(axes.flat, start=1):
    da = ds_month.sel(month=i)   
    im = da.plot(
        ax=ax,
        x="lat",
        y="plev",
        add_colorbar=False
    )
    ax.set_title(f"Month {i}")
    ax.invert_yaxis()
    ax.set_yscale("log") 


plt.tight_layout()
plt.savefig('multiplot_mask_first10y.pdf')
```

```python
ds_zonal = mask.mean(dim="lon")
ds_month = ds_zonal.groupby("time.month").mean("time")


fig, axes = plt.subplots(3, 4, figsize=(16,10), sharex=True, sharey=True)

for i, ax in enumerate(axes.flat, start=1):
    da = ds_month.sel(month=i)   
    im = da.plot(
        ax=ax,
        x="lat",
        y="plev",
        add_colorbar=False
    )
    ax.set_title(f"Month {i}")
    ax.invert_yaxis()


plt.tight_layout()
plt.savefig('multiplot_mask.pdf')
```

```python
da=mask.mean("time")
fig, axes = plt.subplots(3, 6, figsize=(14, 8), sharex=True, sharey=True)

for i, ax in enumerate(axes.flat):
    m = i 
    
    da.sel(plev=m).plot(
        ax=ax,
        x="lon",
        y="lat",
        yincrease=False,      # pressione verso il basso (standard atmosferico)
        add_colorbar=False
    )
    ax.set_title(f"Plev {m}")

fig.tight_layout()

# colorbar unica
cbar = fig.colorbar(ax.collections[0], ax=axes, orientation="vertical")
fig.savefig('multiplot_mask_lev.pdf')
```
